```python
df1 = df1.drop(df1.index[-1])
```

```python
df1 = df1.drop(['Of which: Competitive bids',': Non-competitive bids','Bid-to-Cover Ratio','Rollover / Redemptions','New Borrowing/Net Repayment'], axis =1)
```

```python
# Créez une fonction pour nettoyer les valeurs numériques
def remove_commas(value):
    if isinstance(value, str):
        # Supprimez les virgules et les espaces, puis remplacez les virgules par des points
        return float(value.replace(',', '').replace(' ', '').replace(',', '.'))
    else:
        return value

# Appliquez la fonction aux colonnes numériques
columns_to_clean = ['Amount Offered (Kshs. M)','Bids Received (Kshs. M)','Total Amount Accepted (Kshs. M)']

for column in columns_to_clean:
    df1[column] = df1[column].apply(remove_commas)
```

```python
df1['Due Date'] = pd.to_datetime(df1['Due Date'], format ='%d/%m/%Y').dt.strftime('%d/%m/%Y')
```

```python
# Liste des colonnes à convertir en float
columns_to_convert = ['Market Weighted Average Interest', 'Weighted Average Interest Rate of', 'Price per Kshs 100 at average interest']

for column in columns_to_convert:
    df1[column] = df1[column].str.replace('%', '').astype(float)
```

```python
df1
```

```python
# Liste des colonnes à multiplier par 1,000,000
columns_to_multiply = ['Amount Offered (Kshs. M)', 'Bids Received (Kshs. M)', 'Total Amount Accepted (Kshs. M)']

# Définissez une fonction de multiplication
def multiply_by_million(value):
    return value * 1000000

# Appliquez la fonction de multiplication aux colonnes spécifiées
for column in columns_to_multiply:
    df1[column] = df1[column].apply(multiply_by_million)
```
