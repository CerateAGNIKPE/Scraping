```python
from tabula.io import read_pdf
import pandas as pd
import tabula
from tabulate import tabulate
```

```python
df = tabula.io.read_pdf('1037051525_AUGUST 2023 NEW FXD1-2023-02 & RE-OPENED FXD1-2023-05 DATED 21-08-2023.pdf', pages = 'all', encoding ='utf-8')
df = df[0]
```

```python
df
```

```python
df.set_axis(df.iloc[1], axis=1, inplace=True)
```

```python
df
```

**Supprimer les lignes inutil**

```python
df = df.drop([0,1,3,5,7,8,9,14,15,16,17,18])
```

```python
df
```

```python
df.set_index("TENOR", inplace=True)
```

```python
df1 =df.T
```

```python
df1
```

```python
df1 = df1.drop(df1.index[-1])
```

```python
# Créez une fonction pour nettoyer les valeurs numériques
def remove_commas(value):
    if isinstance(value, str):
        # Supprimez les virgules et les espaces, puis remplacez les virgules par des points
        return float(value.replace(',', '').replace(' ', '').replace(',', '.'))
    else:
        return value

# Appliquez la fonction aux colonnes numériques
columns_to_clean = ['Amount Accepted (Kshs. M)','Total bids Received at cost (Kshs. M)']

for column in columns_to_clean:
    df1[column] = df1[column].apply(remove_commas)
```

```python
df1['Due Dates'] = pd.to_datetime(df1['Due Dates'], format ='%d/%m/%Y').dt.strftime('%d/%m/%Y')
```

```python
columns_to_convert = ['Market Weighted Average Rate (%)','Weighted Average Rate of Accepted Bids (%)','Price per Kshs 100 at average yield','Coupon Rate (%)']

for column in columns_to_convert:
    df1[column] = df1[column].astype(float)
```

**EXPORTATION DES DONNEES**

```python
# Spécifiez le nom du fichier Excel de destination
nom_fichier_excel = 'RE_open_kenya_scrap.xlsx'

# Exportez le DataFrame vers Excel
df1.to_excel(nom_fichier_excel, index=True)
```
