```python
from tabula.io import read_pdf
import pandas as pd
import tabula
from tabulate import tabulate
```

```python
df = tabula.io.read_pdf('Summary of FGN Bond Auction Results for February, 2023.pdf', pages = 'all', encoding ='utf-8')
df = df[0]
```

```python
df
```

```python
df.set_index("Unnamed: 0", inplace=True)
```

```python
df
```

```python
df1 =df.T
df1 
```

```python
df1.dtypes
```

**Conversion des colonnes**

```python
df1.index = df1.index.str.replace(' ', '')
```

```python
# Convertissez les colonnes de date en format "DD/MM/YYYY"
date_columns = ["Auction Date:", "Settlement Date:", "Maturity Date:"]

for col in date_columns:
    df1[col] = pd.to_datetime(df1[col], format='%B %d, %Y').dt.strftime('%d/%m/%Y')

# Affichez le DataFrame résultant
```

```python
# Fonction pour nettoyer les colonnes
def clean_amount(column):
    cleaned = column.str.replace('₦', '').str.replace(' billion', '').str.strip()
    return cleaned

# Appliquez la fonction de nettoyage aux colonnes appropriées
columns_to_clean = ["Amount Offered:", "Subscription:", "Amount Allotted:"]
df1[columns_to_clean] = df1[columns_to_clean].apply(clean_amount)
```

```python
# Liste des colonnes à multiplier par 100000
columns_to_multiply = ["Amount Offered:", "Total Bids:", "Successful Bids:", "Subscription:", "Amount Allotted:"]

df1[columns_to_multiply] = df1[columns_to_multiply].astype(float)
# Multipliez seulement les colonnes spécifiées par 100000
df1[columns_to_multiply] = df1[columns_to_multiply] * 1000000
```

```python
# Nettoyez les valeurs en supprimant le signe de pourcentage et convertissez-les en float
df1['Marginal Rates:'] = df1["Marginal Rates:"].str.replace('%', '').astype(float)
```

```python
df1['Interest Rate'] = df1.index.str.slice(0, 5)
```

```python
df1['Interest Rate'] = df1['Interest Rate'].astype(float)
```
