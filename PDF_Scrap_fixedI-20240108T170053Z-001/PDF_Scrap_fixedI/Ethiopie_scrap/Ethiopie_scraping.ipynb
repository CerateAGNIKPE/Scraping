```python
from tabula.io import read_pdf
import pandas as pd
import tabula
from tabulate import tabulate
```

```python
df = tabula.io.read_pdf('aug-23-2023.pdf', pages = 'all', encoding ='utf-8')
df = df[0].fillna('')
```

```python
df.set_index("Unnamed: 0", inplace=True)
```

```python
df1 =df.T
```

```python
df1 = df1.drop(df1.index[-1])
```

```python
df1
```

```python
import pandas as pd

# Supposons que vous ayez déjà un DataFrame df

# Créez une fonction pour nettoyer les valeurs numériques
def remove_commas(value):
    if isinstance(value, str):
        # Supprimez les virgules et les espaces, puis remplacez les virgules par des points
        return float(value.replace(',', '').replace(' ', '').replace(',', '.'))
    else:
        return value

# Appliquez la fonction aux colonnes numériques
columns_to_clean = ["Amount Offered (ETB Millions)", "Bids Received (ETB Millions)", "Total Amount Accepted (ETB Millions)", 
                    "Competitive Bids"]

for column in columns_to_clean:
    df1[column] = df1[column].apply(remove_commas)

# Maintenant, les virgules ont été enlevées des valeurs numériques.
```

```python
# Colonnes avec des espaces à supprimer
columns_with_spaces = ["Weighted Average Price (Per Birr 100)", "Weighted Average Yield (Annual in %)"]

# Utilisez str.replace() pour supprimer les espaces
for column in columns_with_spaces:
    df1[column] = df1[column].str.replace(' ', '')
```

```python
# Colonnes à multiplier
columns_to_multiply = ["Amount Offered (ETB Millions)", "Bids Received (ETB Millions)"]

# Appliquer la multiplication par 100000 à ces colonnes
df1[columns_to_multiply] = df1[columns_to_multiply].apply(lambda x: x * 1000000)
```

```python
columns_to_convert = ["Cut Off Price (Per Birr 100)", "Cut Off Yield (Annual in %)", "Weighted Average Price (Per Birr 100)", "Weighted Average Yield (Annual in %)"]

# Utilisez pd.to_numeric pour convertir les colonnes en float
for column in columns_to_convert:
    df1[column] = pd.to_numeric(df1[column], errors='coerce')
```

```python
df1['Maturity Date (DD/MM/YYYY)'] = pd.to_datetime(df1['Maturity Date (DD/MM/YYYY)'], format='%d-%b-%y')
```

```python
df1['Maturity Date (DD/MM/YYYY)'] = df1['Maturity Date (DD/MM/YYYY)'].dt.strftime('%d/%m/%Y')
```
