```python
import pandas as pd
```

```python
df = pd.read_excel("Ghana.xlsx")
```

```python
df
```

```python
new_df = df.iloc[16:19, [0, 1, 2, 3, 4, 5, 6, 7, 8]]

# Renommez les colonnes du nouveau DataFrame
new_df.columns = ['ISIN', 'SECURITIES', 'BIDS (AMT) TENDERED GH¢ (M)', 'BIDS (AMT) ACCEPTED GH¢ (M)', 'Range', 'Discount', 'Interest', 'Discount Rate', 'Interest Rate']

# Affichez le nouveau DataFrame
new_df
```

```python
columns1 = ['BIDS (AMT) TENDERED GH¢ (M)', 'BIDS (AMT) ACCEPTED GH¢ (M)']

for column in columns1:
    new_df[column] = new_df[column].str.replace('GH¢', '', regex=True)
```

```python
# Supprimez les virgules et convertissez en float
for column in columns1:
    new_df[column] = pd.to_numeric(new_df[column].str.replace(',', ''), errors='coerce')
```

```python
# Liste des colonnes à multiplier par 1,000,000
columns_to_multiply = ['BIDS (AMT) TENDERED GH¢ (M)', 'BIDS (AMT) ACCEPTED GH¢ (M)']

# Multipliez les colonnes par 1,000,000
new_df[columns_to_multiply] = new_df[columns_to_multiply] * 1000000
```

```python
# Liste des colonnes à convertir en nombres flottants
columns_to_convert = ['Discount Rate', 'Interest Rate']

# Convertir les colonnes en nombres flottants
new_df[columns_to_convert] = new_df[columns_to_convert].apply(lambda x: pd.to_numeric(x, errors='coerce'))
```

```python
new_df
```

```python
# Spécifiez le nom du fichier Excel de destination
nom_fichier_excel = 'Ghana_scrap.xlsx'

# Exportez le DataFrame vers Excel
new_df.to_excel(nom_fichier_excel, index=True)
```
